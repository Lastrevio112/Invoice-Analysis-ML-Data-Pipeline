```python
from google.cloud import bigquery
import os
```

```python
# This should work both locally and through GitHub actions, although we likely will never run it through GitHub actions
creds_path = os.environ.get("GOOGLE_APPLICATION_CREDENTIALS")

if creds_path:
    os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = creds_path
```

```python
client = bigquery.Client(project="invoiceanalysispipeline")
```

```python
client.query("""
    CREATE TABLE IF NOT EXISTS `invoiceanalysispipeline.silver.ds1_invoice_lines` (
        unique_invoice_line_no STRING,
        unique_invoice_no STRING,
        client STRING,
        client_tax_id STRING,
        iban STRING,
        invoice_date DATE,
        invoice_no STRING,
        seller STRING,
        seller_tax_id STRING,
        total_net_worth FLOAT64,
        total_vat FLOAT64,
        item_desc STRING,
        item_net_price FLOAT64,
        item_qty INT64,
        item_vat FLOAT64
    )
""").result()
```
