```python
from google.cloud import bigquery
import os
```

```python
# This should work both locally and through GitHub actions, although we likely will never run it through GitHub actions
creds_path = os.environ.get("GOOGLE_APPLICATION_CREDENTIALS")

if creds_path:
    os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = creds_path
```

```python
client = bigquery.Client(project="invoiceanalysispipeline")
```

```python
client.query("""
    CREATE TABLE IF NOT EXISTS `invoiceanalysispipeline.invoices_gold.dim_vendor` (
        data_source_id STRING,
        vendor_no_id STRING,    --Unique per entire table
        vendor_no STRING,       --Unique per data source
        vendor_desc STRING,
        vendor_address STRING
    )
""").result()
```

```python
client.query("""
    CREATE TABLE IF NOT EXISTS `invoiceanalysispipeline.invoices_gold.dim_buyer` (
        data_source_id STRING,
        buyer_no_id STRING,    --Unique per entire table
        buyer_no STRING,       --Unique per data source
        buyer_desc STRING,
        buyer_address STRING
    )
""").result()
```

```python
client.query("""
    CREATE TABLE IF NOT EXISTS `invoiceanalysispipeline.invoices_gold.fact_invoice_line` (
        row_id STRING,                  --Unique per entire table
        data_source_id STRING,
        unique_invoice_line_no STRING,  --Unique per data source in this table
        unique_invoice_no STRING,
        vendor_no_id STRING,
        buyer_no_id STRING,
        spend FLOAT64, 
        iban STRING,
        invoice_date DATE,
        invoice_no STRING,
        item_desc STRING,
        item_net_price FLOAT64,
        item_qty INT64    
    )
""").result()
```
