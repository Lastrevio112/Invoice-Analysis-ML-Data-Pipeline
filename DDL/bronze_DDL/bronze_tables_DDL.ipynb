Yes, I'm using notebooks for DDL, since these are supposed to be run only once unless a schema changes. 

Therefore notebooks are superior, since you can run them cell by cell and you have more control over it - DDL should never be run automatically by an orchestrator in the context of this project. 

Can't change my mind about notebooks still having a place in production.

```python
from google.cloud import bigquery
import os
```

```python
# This should work both locally and through GitHub actions, although we likely will never run it through GitHub actions
creds_path = os.environ.get("GOOGLE_APPLICATION_CREDENTIALS")

if creds_path:
    os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = creds_path
```

```python
client = bigquery.Client(project="invoiceanalysispipeline")
```

```python
client.query("""
    CREATE TABLE IF NOT EXISTS `invoiceanalysispipeline.bronze.ds_1_raw_json` (
        id STRING,
        raw_json JSON,
        inserted_at TIMESTAMP,
        exclude_from_silver BOOLEAN,     --This makes backfilling easier, as it's not a good practice to delete from bronze, so we add a flag for wrong/bad data that should not reach the silver in the next dbt run
        file_name_in_R2 STRING
    )
""").result()
```

```python
client.query("""
    CREATE TABLE IF NOT EXISTS `invoiceanalysispipeline.bronze.ds_2_raw_json` (
        id STRING,
        raw_json JSON,
        inserted_at TIMESTAMP,
        exclude_from_silver BOOLEAN,
        file_name_in_R2 STRING
    )
""").result()
```
